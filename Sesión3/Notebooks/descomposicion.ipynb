# Ejemplo de descomposición de una serie de tiempo

```python
# import de liberías
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python

# Cargar datos de tiempo
# Aquí utilizamos datos de ejemplo disponibles en pandas
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url)

# Renombramos la columna y convertimos la columna de fecha
data.columns = ['Month', 'Passengers']
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
```

## Descomposición con modelo aditivo

```python

# Descomposición de la serie temporal con modelo aditivo
result = seasonal_decompose(data['Passengers'], model='additive')
result
```

```python

# Graficar la descomposición
fig = result.plot()
plt.suptitle('Descomposición de la Serie Temporal de Pasajeros de Aerolíneas con modelo aditivo', fontsize=12)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Ajusta el título para evitar solapamiento
plt.show()
```

## Descomposición con modelo multiplicativo

```python
# Descomposición de la serie temporal con modelo aditivo
result = seasonal_decompose(data['Passengers'], model='multiplicative')
result
```

```python
# Graficar la descomposición
fig = result.plot()
plt.suptitle('Descomposición de la Serie Temporal de Pasajeros de Aerolíneas con modelo multiplicativo', fontsize=12)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Ajusta el título para evitar solapamiento
plt.show()
```
