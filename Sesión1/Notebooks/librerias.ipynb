## Limpieza de strings con expresiones regulares: re
La librería re nos permite acceder a todo el universo de expresiones regulares para manipulación y limpieza de cadena de caracteres

```python
# importamos la librería
import re
```

```python
# buscar un patrón en una cadena
nombre = "Pablo Alonso Sandoval Mendoza tiene 28 años"
edad = re.search(r'\d+', nombre) # r'\d+' es la expresión regular para buscar números 
edad.group()
```

```python
# buscar al inicio de la cadena
a = re.match('P', nombre) # si buscamos cualquier otra cosa nos regresa None
a.group()
```

```python
# buscar en toda la cadena
al = re.findall('al', nombre)
al
```

```python
# generar un iterador con los matches con el patrón
al2 = re.finditer('al', nombre)
print(al2)
for l in al2:
    print(l.group())
```

```python
# substituir una parte del string
nombre_mal = "Pablo Alonzo Sandoval Mendoza"
re.sub("Alonzo", "Alonso", nombre_mal)
```

```python
# dividir una cadena de caracteres por un separador
nombre_guion = "Pablo-Alonso-Sandoval-Mendoza"
re.split("-", nombre_guion) # regresa un listado con varios elementos, dependiendo cuántas separaciones hizo
```

```python
# construir una función con un regex para utilizarlo
pattern = re.compile(r'\d+') # inicializamos un complie para búsqueda de números
print(pattern.findall("Andrea Ascencio tiene 29 años"))
print(pattern.findall("Alonso Sandoval tiene 28 años"))
```

## Manipulación de variables temporales con datetime y time

```python
# import de los módulos
from datetime import datetime
```

```python
# timestamp de ahorita
datetime.now()
```

```python
# crear un objeto tipo datetime
datetime(2024,9,2,19,0,0)
```

```python
# dar formato específico
ahorita = datetime.now()
print(ahorita.strftime("%Y-%m")) # para recordar la función pienso en string format time -> str f time
# existen diversos formatos para mostrar la fecha o las horas
print(ahorita.strftime("%Y-%B %H:%M")) # mostrando el mes en nombre y la hora
```

```python
# convertir un string a objeto tipo fecha
str_fecha = "2024-10-08"
datetime.strptime(str_fecha, "%Y-%m-%d") # recuerdo el nombre de la función por string parse time -> str p time
# en esta función hay que mandar el formato real del string y no el formato deseado
```

```python
# obtener solo la fecha de hoy
# camino 1
from datetime import date
hoy = date.today()
print(hoy)
# camino 2
ahorita = datetime.now()
hoy = ahorita.strftime("%Y-%m-%d")
print(hoy)
```

```python
# diferencia entre fechas
fecha1= date(2024,9,2)
fecha2 = date(2024,8,1)
delta = fecha1- fecha2
# diferencia en días
print(delta.days)
```

```python
# podemos obtener cada componente de la fecha haciendo uso de los atributos del objeto tipo date
print(fecha1.year)
print(fecha1.month)
print(fecha1.day)
```

```python
# operaciones con dias
from datetime import timedelta
ahorita = datetime.now()
ahorita_mas_30dias = ahorita + timedelta(days=30)
print(ahorita_mas_30dias)
```

## Traceback
traceback es una librería úti para manejo de errores (excepciones) y de la mano con un código dentro de bloques try catch puedes construir scripts robustos que aunque fallen en ejecutar ciertas partes, se termine de ejecutar o por lo menos tener un rastreo útil del error

```python
# import de librería
import traceback
import pandas as pd # utilizaremos pandas para nuestro ejemplo
```

```python
# hagamos una función que dependiendo el input podría regresar un tipo de error (es una función mal hecha para producir una excepción)
def separa_nombres(nombre):
    '''
    Función para dividir una cadena de caracter en todos sus valores
    INPUT: 
    - nombre: [str] con nombre a separar con el separador "-"
    OUTPUT:
    prints de los 4 nombres # aquí es donde va a poder existir el error si es que no hay 4 valores
    '''
    try:
        nombres = pd.DataFrame()
        nombres["nombre"] = nombre # guardamos en una columna el nombre
        sepa = nombre.split("-")
        for i in range(4): 
            print(sepa[i])
    except Exception as e: # la parte de Exception as e es opcional  yestamos indicando que queremos cachar la excepción
        # en este bloque ponemos qué queremos que haga cuando tenga un error
        print("Error separando el nombre {}:\n{}".format(nombre,traceback.format_exc()))
    else:
        # en este bloque ponemos lo que queremos que ponga sólo en caso de que no haya tenido error
        print("Corrí correctamente")
    finally:
        # En este bloque ponemos lo que queremos que siga ejecutando sin importar si hubo error o no
        print("Hello world")
    
```

```python
# hablemos a la función con distintos nombres
separa_nombres("Pablo-Alonso-Sandoval-Mendoza")
```

```python
# si hablamos a la función con menos nombres se genera la excepción
separa_nombres("Andrea-Ascencio-Dueñas")
# aunque hubo error imprime el hello world por estar en el bloque finally
```

## Visualizaciones con matplotlib y seaborn
Ahora generaremos visualizaciones de datos utilizando matplotlib y seaborn. Puedes escoger una de estas 2 para tus gráficos, en general seaborn parecen más bonitos pero desde matplotlib puedes ahcer varias cosas

```python
# import de las librerías
import pandas as pd
import seaborn as sns # al igual que pandas y numpy, la abreviación normalmente usada para seaborn es sns
import matplotlib.pyplot as plt # igualmente en este caso, la abreviación normalmente usada es plt
```

```python
# utilizaremos el dataset de iris para generar visualizaciones
iris = sns.load_dataset('iris')
iris
```

```python
# generar un scatter plot
# con seaborn. Estamos utilizando también matplotlib para generar los márgenes y cambiar más elementos del gráfico
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# pairplot para comparar todas las variables, con hue especificamos que queremos que coloree distinto
sns.pairplot(iris, hue='species', palette='Set1')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)  # Adjust title position
plt.show()
```

```python
# boxplot del sepal_length por especie
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_length', palette='Set1')
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris, x='species', y='petal_length', palette='Set1')
plt.title('Violin Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# heatmap de matriz de correlación entre variables
plt.figure(figsize=(10, 6))
correlation_matrix = iris[['sepal_length','sepal_width','petal_length','petal_width']].corr() # los objetos tipo dataframe tienen el método corr() para calcular la correlación
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Iris Dataset Features')
plt.show()
```

```python
# histogramas
plt.figure(figsize=(10, 6))
sns.histplot(iris['petal_length'], kde=True, bins=20, color='purple')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# agregar facet grid para agregar más dimensiones a los gráficos
# aqui cambia un poco la sintáxis ya que ahora tenemos que generar primero el facetgrid y luego mapearle algún tipo de gráfico
g = sns.FacetGrid(iris, col='species', height=4)
g.map(sns.histplot, 'sepal_length', kde=True, bins=10)
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Distribution of Sepal Length by Species')
plt.show()
```
