# Ejemplo de estimaciones para un evento probabilístico
Utilizaremos información de un dataset para Machine Learning con información de poblaciones. El dataset lo puedes descargar aquí: [Adult](https://archive.ics.uci.edu/dataset/2/adult)

```python
# import de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuración de estilo
sns.set(style="whitegrid")

# Definición de las columnas basadas en el conjunto de datos Adult
columnas = ['Edad', 'Trabajo', 'income', 'NivelEducativo', 'HorasPorSemana', 'Estado civil'
            , 'sector', 'Estado familia', 'Raza','Género','numero', 'años','años trabajo','Origen', 'Ganancia']
```

```python
#url = url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
df = pd.read_csv("datasets/adult/adult.data", header=None, names=columnas, na_values=' ?', skipinitialspace=True)
```

```python
df
```

```python
# calcularemos estimadores sobre la métrica edad
df = df[['Edad']].dropna()

# Calcular estimadores
media_edad = df['Edad'].mean()
mediana_edad = df['Edad'].median()
varianza_edad = df['Edad'].var()
desviacion_estandar_edad = df['Edad'].std()

# Mostrar resultados
print(f"Media de la edad: {media_edad:.2f}")
print(f"Mediana de la edad: {mediana_edad:.2f}")
print(f"Varianza de la edad: {varianza_edad:.2f}")
print(f"Desviación estándar de la edad: {desviacion_estandar_edad:.2f}")

# Graficar la distribución de edades
plt.figure(figsize=(10, 6))
sns.histplot(df['Edad'], bins=30, kde=True, color='blue', stat="density")
plt.title('Distribución de Edades en el Dataset Adult')
plt.xlabel('Edad')
plt.ylabel('Densidad de Probabilidad')
plt.axvline(media_edad, color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana_edad, color='orange', linestyle='dashed', linewidth=2, label='Mediana')
plt.legend()
plt.grid(True)
plt.show()

```

conceptualmente, si estuvieramos asumiendo que los datos se distribuyen normal necesitaríamos estimar el parámetro mu y el parámetro sigma.
En este caso estaríamos haciendo una estimación del parámetro mu con la media muestral (38.58) y una estimación del parámetro sigma con la desviación estándar muestral (13.64).  
Usando estos 2 parámetros podríamos simular datos pertenecientes a esta población con el modelo de una distribución Normal(38.58,13.64)
