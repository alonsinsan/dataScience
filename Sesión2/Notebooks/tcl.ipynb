# Ejemplo con Teorema Central del Límite
En este ejemplo simularemos poblaciones de una distribución no normal, calculando la media sobre estos datos y haciendo su histograma verificaremos el teorema central del límite

```python
# import de librerías
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Configuración de estilo
sns.set(style="whitegrid")

# Parámetros
np.random.seed(42)  # Para reproducibilidad
n_muestras = 10000   # Número de muestras
n_tamano_muestra = 30  # Tamaño de cada muestra
bajo = 0            # Límite inferior de la distribución uniforme
alto = 100         # Límite superior de la distribución uniforme
```

```python
# Generar muestras de una distribución uniforme
muestras = np.random.uniform(bajo, alto, (n_muestras, n_tamano_muestra)) # generamos n_muestras de tamaño n_tamano_muestra

# Calcular la media de cada muestra
medias_muestras = np.mean(muestras, axis=1)
```

```python
# graficamos las medias de la población con un histograma
plt.figure(figsize=(12, 6))
# Histograma de las medias de las muestras
sns.histplot(medias_muestras, bins=30, kde=True, color='blue', stat='density')
plt.title('Distribución de las Medias de Muestras (Distribución Uniforme)')
plt.xlabel('Media de la Muestra')
plt.ylabel('Densidad de Probabilidad')
# agreguemos una normal teórica
mu, sigma = np.mean(medias_muestras), np.std(medias_muestras)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r--', label='Distribución Normal Teórica')
plt.legend()
plt.grid(True)
plt.show()
```

conceptualmente, si tenemos un estimador para una población (sin saber la distribución "original" de ese estimador) y tenemos acceso a distintas mediciones de ese estimador. el teorema central del límite dice que ese estimador tendrá una distribución normal por lo que podremos generar simulaciones sobre el valor de ese parámetro o generar un intervalo de confianza para el valor de ese parámetro a través de los valores para una distribución normal
