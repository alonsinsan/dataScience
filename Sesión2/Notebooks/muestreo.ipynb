# Notebook de ejemplo para muestreo probabilístico

```python
# import de librerías a utilizar
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, binom, poisson # dentro de scipy en el sub-módulo stats se encuentran las funciones de probabilidad
```

```python
# Establecer el estilo para las gráficas de las funciones
sns.set(style="whitegrid")

# Función para graficar la distribución normal
def plot_normal_distribution(mu, sigma):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = norm.pdf(x, mu, sigma)
    
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Distribución Normal', color='blue')
    plt.title('Distribución Normal (mu={}, sigma={})'.format(mu, sigma))
    plt.xlabel('Valores')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True)
    plt.show()

# Función para graficar la distribución normal
def plot_binomial_distribution(n, p):
    x = np.arange(0, n+1)
    y = binom.pmf(x, n, p)
    
    plt.figure(figsize=(10, 5))
    plt.bar(x, y, label='Distribución Binomial', color='orange')
    plt.title('Distribución Binomial (n={}, p={})'.format(n, p))
    plt.xlabel('Número de Éxitos')
    plt.ylabel('Probabilidad')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_poisson_distribution(lam, size=1000):
    # Generar datos
    data = np.random.poisson(lam, size)
    
    # Graficar
    plt.figure(figsize=(10, 5))
    sns.histplot(data, bins=30, kde=False, color='green', stat="density", label='Distribución de Poisson')
    plt.title('Distribución de Poisson (lambda={})'.format(lam))
    plt.xlabel('Número de Eventos')
    plt.ylabel('Densidad de Probabilidad')
    
    # Graficar la función de densidad
    x = np.arange(0, max(data)+1)
    y = poisson.pmf(x, lam)
    plt.plot(x, y, 'ro-', label='Densidad de Probabilidad Teórica', markersize=5)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# ejemplo de muestreo para la distribución normal
mu = 0 # parámetro de media
sigma = 1 # parámetro de desviación estándar
plot_normal_distribution(mu, sigma) # cómo mu=0 y sd = 1 estamos graficando la normal estándar
# NOTA: todas las normales con parámetros mu y sd podemos transformarlas a una distribución normal (a este proceso se le llama estandarización)
```

```python
# ejemplo de muestreo para la distribución binomial
n = 10
p = 0.5
plot_binomial_distribution(n,p) # estamos haciendo 10 experimentos con probabilidad 1/2 de éxito
```

```python
# ejemplo de muestreo de distribución poisson
lam = 5
plot_poisson_distribution(lam) # estamos modelando un evento probabilístco con media 1/5 de ocurrencia
```
